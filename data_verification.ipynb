```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('bu_gym_entries_spring.csv')
```

```python
# Convert the 'timestamp' column to datetime format
df['date'] = pd.to_datetime(df['timestamp']).dt.date

# Count the number of swipes per day
swipes_per_day = df['date'].value_counts().sort_index()

# Plot the swipes per day
swipes_per_day.plot(kind='line', figsize=(12, 6), title='Swipes Per Day')
plt.xlabel('Date')
plt.ylabel('Number of Swipes')
plt.grid()
plt.show()
```
